```python
!pip install geopy folium
import folium
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import time

# 1. Khởi tạo Geocoder (Kiến thức Slide 32, 56)
geolocator = Nominatim(user_agent="bai_tap_23_2")

# 2. Xác định điểm trung tâm (Ví dụ: Chợ Bến Thành)
center_address = "Chợ Bến Thành, Quận 1, Thành phố Hồ Chí Minh"
center_loc = geolocator.geocode(center_address)
center_coords = (center_loc.latitude, center_loc.longitude)

# 3. Danh sách 10 địa chỉ tại TP.HCM (Giả lập theo yêu cầu đề bài)
addresses = [
    "Dinh Độc Lập, Quận 1",
    "Nhà thờ Đức Bà, Quận 1",
    "Bưu điện Trung tâm Sài Gòn, Quận 1",
    "Bảo tàng Chứng tích Chiến tranh, Quận 3",
    "Nhà hát Thành phố, Quận 1",
    "Tòa nhà Bitexco, Quận 1",
    "Landmark 81, Quận Bình Thạnh",
    "Hồ Con Rùa, Quận 3",
    "Công viên Tao Đàn, Quận 1",
    "Bảo tàng Thành phố Hồ Chí Minh, Quận 1"
]

# 4. Khởi tạo bản đồ tại điểm trung tâm (Kiến thức Slide 37)
m = folium.Map(location=[center_coords[0], center_coords[1]], zoom_start=14)

# Thêm Marker cho điểm trung tâm (Kiến thức Slide 38)
folium.Marker(
    center_coords,
    popup=f"TRUNG TÂM: {center_address}",
    icon=folium.Icon(color='red', icon='star')
).add_to(m)

# 5. Duyệt danh sách địa chỉ, Geocoding và tính khoảng cách (Kiến thức Slide 33, 35, 57)
print(f"BẮT ĐẦU XỬ LÝ KHOẢNG CÁCH ĐẾN: {center_address}\n" + "-"*50)

for addr in addresses:
    try:
        # Geocoding: Chuyển địa chỉ sang tọa độ
        location = geolocator.geocode(addr)
        if location:
            point_coords = (location.latitude, location.longitude)

            # Tính khoảng cách geodesic (Kiến thức Slide 33)
            dist = geodesic(center_coords, point_coords).km

            # Hiển thị kết quả ra console
            print(f"- {addr}: {dist:.2f} km")

            # Thêm Marker cho địa điểm (Kiến thức Slide 38)
            folium.Marker(
                point_coords,
                popup=f"{addr}<br>Khoảng cách: {dist:.2f} km",
                tooltip=addr,
                icon=folium.Icon(color='blue')
            ).add_to(m)

            # Vẽ đường nối để thể hiện mối quan hệ không gian (Kiến thức Slide 39, 57)
            folium.PolyLine(
                locations=[center_coords, point_coords],
                color='green',
                weight=2,
                opacity=0.6,
                dash_array='5, 5' # Tạo đường đứt nét
            ).add_to(m)

        # Nghỉ một chút để tránh bị giới hạn API Nominatim
        time.sleep(1)
    except Exception as e:
        print(f"Lỗi khi xử lý địa chỉ {addr}: {e}")

# 6. Hiển thị bản đồ
m
```
