```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
delivery = ctrl.Antecedent(np.arange(0, 25, 1), 'delivery')

quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')

cost = ctrl.Antecedent(np.arange(0, 11, 1), 'cost')

suitability = ctrl.Consequent(np.arange(0, 101, 1), 'suitability')

delivery['fast'] = fuzz.trimf(delivery.universe, [0, 0, 10])
delivery['moderate'] = fuzz.trimf(delivery.universe, [5, 12, 18])
delivery['slow'] = fuzz.trimf(delivery.universe, [15, 24, 24])

quality['poor'] = fuzz.trimf(quality.universe, [0, 0, 5])
quality['average'] = fuzz.trimf(quality.universe, [3, 5, 7])
quality['good'] = fuzz.trimf(quality.universe, [5, 10, 10])

cost['low'] = fuzz.trimf(cost.universe, [0, 0, 5])
cost['medium'] = fuzz.trimf(cost.universe, [3, 5, 7])
cost['high'] = fuzz.trimf(cost.universe, [5, 10, 10])


suitability['low'] = fuzz.trimf(suitability.universe, [0, 0, 50])
suitability['medium'] = fuzz.trimf(suitability.universe, [25, 50, 75])
suitability['high'] = fuzz.trimf(suitability.universe, [50, 100, 100])
```

```python

rule1 = ctrl.Rule(delivery['fast'] & quality['good'] & cost['low'], suitability['high'])
rule2 = ctrl.Rule(quality['poor'] | delivery['slow'], suitability['low'])
rule3 = ctrl.Rule(quality['average'] & cost['medium'], suitability['medium'])
rule4 = ctrl.Rule(delivery['fast'] & quality['good'] & cost['high'], suitability['medium'])


evaluation_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
evaluation = ctrl.ControlSystemSimulation(evaluation_ctrl)

evaluation.input['delivery'] = 3
evaluation.input['quality'] = 8
evaluation.input['cost'] = 7


evaluation.compute()


print(f"Điểm phù hợp của nhà cung cấp: {evaluation.output['suitability']:.2f}%")
```
